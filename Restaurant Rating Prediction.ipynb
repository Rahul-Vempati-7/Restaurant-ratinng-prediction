```python
#### If you want to check whether feature is guassian or normal distributed
#### Q-Q plot

def plot_data(df, feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature], dist='norm', plot=pylab)
    plt.show()
```

```python
#performing feature transformation 
import scipy.stats as stat
import pylab

### BoxCOx Transformation
#The Box-Cox transformation is defined as: 
#T(Y)=(Y exp(λ)−1)/λ where Y is the response variable and λ is the transformation parameter. 
#λ varies from -5 to 5. In the transformation, all values of λ  are considered and the optimal value for a
#given variable is selected.

data['approx_cost(for two people)_boxcox'], parameters = stat.boxcox(data['approx_cost(for two people)'])
plot_data(data, 'approx_cost(for two people)_boxcox')
```

```python
data['votes_boxcox'], parameters = stat.boxcox(data['votes'])
plot_data(data, 'votes_boxcox')
```

```python
data['rate_boxcox'], parameters = stat.boxcox(data['rate'])
plot_data(data, 'rate_boxcox')
```

```python
#using Label encoding to convert for the rest of the columns 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

data.location = le.fit_transform(data.location)
data.rest_type = le.fit_transform(data.rest_type)
data.cuisines = le.fit_transform(data.cuisines)
data['listed_in(type)'] = le.fit_transform(data['listed_in(type)'])
```

```python
data.drop(['online_order','book_table'], axis=1, inplace=True)
```

```python
data.rename(columns={'book_table_Yes':'book_table', 'online_order_Yes':'online_order'}, inplace=True)
```

```python
cols = data.columns
```

```python
#feature scaling
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

data = scaler.fit_transform(data)
```

```python
data = pd.DataFrame(data, columns=[cols])
```

```python
data
```

```python
x = data.drop('rate',axis=1).values
y = data['rate'].values
```

```python
y = y.ravel()
```

```python
#splitting
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=0)
```

```python
Models = ['LinearRegression', 'RandomForest', 'Ridge', 'Lasso', 'DecisionTree', 'KNN', 'AdaBoost', 'SVR']
Scores = []
```

```python
#LinearReg
from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()
lr_model.fit(x_train,y_train)
from sklearn.metrics import r2_score
y_pred=lr_model.predict(x_test)
Scores.append(r2_score(y_test,y_pred))
```

```python
#RandomForest
from sklearn.ensemble import RandomForestRegressor
RF_Model=RandomForestRegressor(n_estimators=650,random_state=245,min_samples_leaf=.0001)
RF_Model.fit(x_train,y_train)
y_predict=RF_Model.predict(x_test)
Scores.append(r2_score(y_test,y_predict))
```

```python
# Ridge
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(x_train,y_train)
y_pred = ridge.predict(x_test)
Scores.append(r2_score(y_test,y_pred))
```

```python
#Lasso
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(x_train,y_train)
lasso.score(x_test,y_test)
y_pred = lasso.predict(x_test)
Scores.append(r2_score(y_test,y_pred))
```

```python
#DecisionTree
from sklearn.tree import DecisionTreeRegressor
DTR = DecisionTreeRegressor(max_depth=35)
DTR.fit(x_train, y_train)
y_pred = DTR.predict(x_test)
Scores.append(r2_score(y_pred,y_test))
```

```python
#KNN
from sklearn.neighbors import KNeighborsRegressor
KNN_Reg = KNeighborsRegressor()
KNN_Reg.fit(x_train,y_train)
y_pred = KNN_Reg.predict(x_test)
Scores.append(r2_score(y_test,y_pred))
```

```python
#AdaBoost
from sklearn.ensemble import AdaBoostRegressor
ADA_Reg = AdaBoostRegressor()
ADA_Reg.fit(x_train,y_train)
y_pred = ADA_Reg.predict(x_test)
Scores.append(r2_score(y_test,y_pred))
```

```python
#SVR
from sklearn.svm import SVR
SVR = SVR()
SVR.fit(x_train,y_train)
y_pred = SVR.predict(x_test)
Scores.append(r2_score(y_test,y_pred))
```

```python
df = pd.DataFrame({'Scores':Scores},index=Models)
df
```
